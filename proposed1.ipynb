# 2. Load Dataset

```python
cached_data_path = join(PATH.cache, 'train_val_test.joblib')
if exists(cached_data_path):
    datas = joblib.load(cached_data_path)
```

# 3. Modeling

```python
train_data, val_data, test_data = datas['train'], datas['val'], datas['test']
target, classes = datas['target'], datas['classes']
```

```python
from sklearn.model_selection import train_test_split
from pycaret.classification import *

_, train_data_sample = train_test_split(train_data, test_size=0.01, stratify=train_data[target], random_state=RANDOM_STATE)
_, val_data_sample = train_test_split(val_data, test_size=0.01, stratify=val_data[target], random_state=RANDOM_STATE)
train_data_sample.shape, val_data_sample.shape

s = setup(train_data_sample, target=target, test_data=val_data_sample, use_gpu=True,
          numeric_features=train_data_sample.columns.drop(target).tolist())
```

```python
best = compare_models(cross_validation=False)
```

```python
plot_model(best, plot='confusion_matrix')
```

- `0` class를 제대로 못 맞추고 있음
- Recall
    - **실제 0인 사람들 중 제대로 0이라고 맞춘 비율: 9% (돈 빌려주고 91% 못 받음)**
    - 실제 1인 사람들 중 제대로 1이라고 맞춘 비율: 73%
    - 실제 2인 사람들 중 제대로 2라고 맞춘 비율: 78%
- Precision
    - 0이라고 예측 중 실제 0인 비율: 
    - 1이라고 예측 중 실제 1인 비율: 
    - 2이라고 예측 중 실제 2인 비율: 

```python
1682 / (48 + 421 + 1682)
```
